1. Вычислите A + B, A - B
 
 $\\
A = \left(
\begin{array}{cccc}
1 & -2 & 4\\
2 & 0 & -1\\
\end{array}
\right), B = \left(
\begin{array}{cccc}
5 & 2 & 3\\
4 & 6 & 2\\
\end{array}
\right)$

```python
A = np.array([[1, -2, 4], [2, 0, -1]])
B = np.array([[5, 2, 3], [4, 6, 2]])
A+B, A-B
```

2. Найдите матрицу С = A - 3B
 
 $\\
A = \left(
\begin{array}{cccc}
1 & 2 \\
2 & -1\\
3 & 0\\
\end{array}
\right), B = \left(
\begin{array}{cccc}
-1 & 1 \\
1 & 2\\
0 & 0\\
\end{array}
\right)$

```python
A = np.array([[1, 2], [2, -1], [3, 0]])
B = np.array([[-1, 1], [1, 2], [0, 0]])
C = A-3*B
C
```

3. Вычислите С = AB
 
 $\\
A = \left(
\begin{array}{cccc}
1 & -1 \\
2 & 0\\
3 & 0\\
\end{array}
\right), B = \left(
\begin{array}{cccc}
1 & 1 \\
2 & 0\\
\end{array}
\right)$

```python
A = np.array([[1, -1], [2, 0], [3, 0]])
B = np.array([[1, 1], [2, 0]])
C = A@B
C
```

4. Вычислите определитель матрицы А

8 & 0 & 2\\

```python
A = np.array([[6, -1, 4], [2, 1, 10], [8, 0, 2]])
np.linalg.det(A)
```

5. Вычислите матрицу, обратную матрице А. И покажите, что она действительно является обратной
 
 $\\
A = \left(
\begin{array}{cccc}
3 & 3 & -1\\
4 & 1 & 3\\
1 & -2 & -2\\
\end{array}
\right)$

```python
A = np.array([[3, 3, -1], [4, 1, 3], [1, -2, -2]])
Ainv = np.linalg.inv(A)
assert(np.allclose(A @ Ainv, np.eye(3)))
assert(np.allclose(Ainv @ A, np.eye(3)))
Ainv
```
