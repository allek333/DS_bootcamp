```python
data = pd.read_csv("melb_data.csv")
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
sns.pairplot(data[["Bedroom2", "Bathroom", "Car", "Price", "Landsize", "Rooms"]], hue="Rooms");
```

```python
sns.scatterplot(x="Landsize", y="Price", data=data);
plt.title("Диаграмма рассеяния");
```

```python
sns.relplot(x="Landsize", y="Price", hue="Rooms", data=data);
plt.title("Диаграмма рассеяния");
```

```python
sns.lineplot(x="YearBuilt", y="Price", data=data[:100]);
plt.title("Линейный график");
```

```python
sns.histplot(data["Price"], bins=100);
plt.title("Гистограмма");
```

```python
sns.boxplot(x="Rooms", y="Price", data=data);
plt.title("Ящик с усами");
```

```python
sns.barplot(x="Rooms", y="Price", data=data);
plt.title("Столбчатая диаграмма");
```

```python
sns.countplot(x="Rooms", data=data);
```

```python
plt.figure(figsize=(10, 10))
sns.heatmap(data.corr(), annot=True);
plt.title("Тепловая карта");
```

```python
sns.regplot(x="Landsize", y="Price", data=data);
plt.title("Линейная регрессия");
```

```python
sns.distplot(data.Price);
plt.title("Функция распределения");
```

```python
sns.jointplot(x="Rooms", y="Price", data=data);
plt.title("График совместного распределения");
```

```python
sns.violinplot(x="Rooms", y="Price", data=data);
plt.title("Скрипичная диаграмма");
```
